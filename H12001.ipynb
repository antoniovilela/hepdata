```python
import json
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams.update(
    {'font.size': 18,
     'font.family': 'sans-serif',
     'legend.fontsize': 14,
     'axes.labelsize': 22,
     'axes.labelpad': 8.0,
     'xtick.labelsize': 14,
     'ytick.labelsize': 14,
     'errorbar.capsize': 4
    }
    )
```

```python
data_table_F2_1 = None
data_table_F2_2 = None
with open( "HEPData/H12001-HEPData-ins539089-Table_1.json" ) as f:
    data_table_F2_1 = json.load( f )
with open( "HEPData/H12001-HEPData-ins539089-Table_2.json" ) as f:
    data_table_F2_2 = json.load( f )
```

```python
from get_df_json import *
df_F2_1 = get_df_json( data_table_F2_1, "F2")
df_F2_1
```

```python
df_F2_2 = get_df_json( data_table_F2_2, "F2")
df_F2_2
```

```python
df_F2 = pd.concat( [ df_F2_1, df_F2_2 ] )
df_F2
```

```python
df_F2.loc[ :, 'tot' ] = np.sqrt( df_F2.loc[ :, 'stat' ] ** 2 + df_F2.loc[ :, 'sys_1' ] ** 2 + df_F2.loc[ :, 'sys_2' ] ** 2 )
df_F2_index = df_F2.set_index( ['x', 'Q2' ] ).sort_index()
df_F2_index
```

```python
arr_group_size = df_F2_index.groupby( ['x', 'Q2'] ).size()
np.sum( arr_group_size != 1 )
```

```python
index_x = df_F2_index.index.get_level_values( 0 ).drop_duplicates()
index_x
```

```python
for x__ in index_x:
    print ( x__, df_F2_index.loc[ x__, [ 'F2', 'tot' ] ] )
```

```python
fig = plt.figure( figsize=(12,10) )
markers = [ 'o', 'v', 's', 'P' ]
for idx__, x__ in enumerate( index_x ):
    df__ = df_F2_index.loc[ x__, [ 'F2', 'tot' ] ]
    val__ = df__.loc[ :, 'F2' ]
    err__ = df__.loc[ :, 'tot' ]
    plt.errorbar( df__.index, val__, yerr=err__, label="x={}".format( x__ ), marker=markers[ idx__ % len(markers) ], fillstyle='none', markersize=12, linestyle='none' )

plt.xlabel( r'$Q^2\,(\rm{GeV}^2)$' )
plt.ylabel( r'$F_2\,(x,Q^2)$' )
plt.xlim( 1e+00, 1e+04 )
plt.ylim( 1e-01, 1e+01 )
plt.xscale( 'log' )
plt.yscale( 'log' )
plt.legend( loc='best' )
```
